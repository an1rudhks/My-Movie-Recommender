🎬 My Movie Recommender: AI Research & Logic

This notebook documents the Research & Development phase of the project. I have build a Content-Based Filtering engine that recommends movies by calculating the mathematical similarity between their metadata tags.

Key Steps:

Data Ingestion: Loading the curated 100-movie dataset.

Text Vectorization: Converting movie tags into numerical data.

Similarity Calculation: Using Cosine Similarity to find the closest matches.

Logic Verification: Testing the engine before deployment to the Streamlit app.

Step 1: Environment Setup and Data Loading

We start by importing Pandas for data manipulation and Scikit-Learn for the machine learning logic. We then load our movies_data.csv file, which contains the movie titles and the pre-processed tags.

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Load the dataset
# Note: '../' is used to go up one folder level from 'notebooks' to find 'data'
df = pd.read_csv('../data/movies_data.csv')

print(f"Total Movies Loaded: {len(df)}")
df.head()
```

Step 2: Vectorization

The CountVectorizer converts the tags column into a numerical matrix where each column represents a word and each row represents a movie's word count.

```python
# Initialize Vectorizer with 5000 common words and remove 'stop words' (e.g., 'and', 'the')
cv = CountVectorizer(max_features=5000, stop_words='english')

# Transform tags into a matrix of numbers
vectors = cv.fit_transform(df['tags']).toarray()

print(f"Vector Matrix created with shape: {vectors.shape}")
```

Step 3: Calculating Cosine Similarity

Calculating the Cosine Similarity between every movie vector. This creates a score between 0 and 1; a score closer to 1 indicates that the movies are highly similar based on their tags.

```python
# Generate the similarity matrix
similarity = cosine_similarity(vectors)

print("Similarity Matrix generated. The 'Brain' of the recommender is ready.")
```

Step 4: Logic Verification

This function mimics the user clicking the "Show Recommendations" button in the final app. It identifies the index of the selected movie, sorts the similarity scores, and retrieves the top 5 closest neighbors.

```python
# Description: This function mimics the user clicking the 'Show Recommendations' button.
# It identifies the movie index, sorts the similarity scores, and prints the top 5 matches.

def get_recommendations(movie_title):
    # 1. Find the index of the user-selected movie
    idx = df[df['title'] == movie_title].index[0]
    
    # 2. Get similarity distances and sort them (Highest similarity first)
    # We use enumerate to keep track of the original movie index during sorting.
    distances = sorted(list(enumerate(similarity[idx])), reverse=True, key=lambda x: x[1])
    
    # 3. Print the top 5 results (Skipping the 0th index, which is the movie itself)
    print(f"Recommendations for '{movie_title}':\n")
    for i in distances[1:6]:
        print(f"- {df.iloc[i[0]].title} (TMDB ID: {df.iloc[i[0]].movie_id})")

# Test run to verify results
get_recommendations('Batman Begins')
```
