**🎬 Movie Recommender System**

**Objective:**  
Build a Content-Based Movie Recommendation System using machine learning.

**Approach:**  
- CountVectorizer → Convert text to numbers  
- Cosine Similarity → Find similar movies  

This logic is used in the Streamlit application.

**📂 Step 1: Load Dataset**

We load movie data containing:
- Title
- Tags (metadata for recommendation)

We also clean missing values for better processing.

```python
import pandas as pd

df = pd.read_csv('../data/movies_data.csv')

df['tags'] = df['tags'].fillna('')
df['title'] = df['title'].astype(str).str.strip()

print("Movies Loaded:", len(df))
```

**🔢 Step 2: Text Vectorization**

We convert movie tags into numerical vectors using CountVectorizer.

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

cv = CountVectorizer(max_features=5000, stop_words='english')
vectors = cv.fit_transform(df['tags']).toarray()

similarity = cosine_similarity(vectors)

print("Model Ready ✔")
```

**🎯 Step 3: Recommendation Engine**

We compute similarity scores and return top 5 most similar movies.

```python
def recommend(movie):
    index = df[df['title'] == movie].index[0]

    distances = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda x: x[1])

    print(f"\nRecommendations for: {movie}\n")

    count = 0
    for i in distances:
        if df.iloc[i[0]].title == movie:
            continue

        print(df.iloc[i[0]].title)

        count += 1
        if count == 5:
            break

recommend("Batman Begins")
```

**✅ Conclusion**

- Built a content-based recommender system
- Used vectorization + cosine similarity
- Verified recommendations before deployment

This system is deployed using Streamlit for real-time use.
