```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Load data using the relative path
df = pd.read_csv('../data/movies_data.csv')

print("Setup Complete! Data loaded successfully.")
df.head()
```

```python
# Initialize the Vectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')

# Transform the 'tags' column into a matrix of numbers
vectors = cv.fit_transform(df['tags']).toarray()

print(f"We now have a matrix of {vectors.shape[0]} movies and {vectors.shape[1]} unique words.")
```

```python
# Calculate Cosine Similarity
similarity = cosine_similarity(vectors)

# Preview the similarity matrix
print(similarity)
```

```python
def get_recommendations(movie_name):
    # Find the row index of the movie
    movie_index = df[df['title'] == movie_name].index[0]
    
    # Get the list of similarity scores for that movie
    distances = similarity[movie_index]
    
    # Sort the movies based on similarity scores (highest first)
    # enumerate() keeps track of the original index
    movie_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
    
    print(f"If you liked '{movie_name}', you might also like:")
    for i in movie_list:
        print(f"- {df.iloc[i[0]].title}")

# Test the function
get_recommendations('Avatar')
```
